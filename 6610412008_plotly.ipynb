```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('https://github.com/jakphunn/5001_plotly/blob/main/country_wise_latest.csv?raw=True')
df.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
# Cleaning data
# Rename the hard to read columns
df = df.rename({"Country/Region" : "Country", "Deaths / 100 Cases" : "Deaths_percentage", "Recovered / 100 Cases" : "Recovered_percentage", "WHO Region" : "Region"}, axis = "columns")
df
```

```python
#drawing bar charts
df1 = df
df1 = df1.groupby("Region").sum()
dfa = df1.assign(Death_percent = lambda x: (x['Deaths'] / x['Confirmed']) * 100)
dfa = dfa.sort_values(by = 'Death_percent', ascending= False)

tt = "percentage of death from corona virus sorted by regions"

fig = px.bar(dfa, x = dfa.index, y = "Death_percent" , color = "Death_percent", title = tt, color_continuous_scale= 'Burgyl', labels = {"Death_percent" : "Death %"})
fig.show()
```
